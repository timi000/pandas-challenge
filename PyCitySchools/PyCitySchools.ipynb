```python
# Dependencies and Setup

import pandas as pd


# File to Load (Remember to Change These)
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Format Floats 
pd.options.display.float_format = '{:,.2f}'.format

# Change Budget to a float 
school_data_complete['budget']= school_data_complete['budget'].astype(float)
                                   
school_data_complete.head()

```

```python
# Store key values into Variables 
Total_School =len(school_data_complete["school_name"].unique())
Total_Students=school_data_complete["student_name"].count()
Total_Budget=(school_data_complete["budget"].unique().sum())
average_math=school_data_complete["math_score"].mean()
average_reading=school_data_complete["reading_score"].mean()
pass_reading=school_data_complete.loc[school_data_complete["reading_score"]>=70, "Student ID"].count()
pass_math=school_data_complete.loc[school_data_complete["math_score"]>=70, "Student ID"].count()
pass_m=round((pass_math/Total_Students)*100,3)
pass_r=round((pass_reading/Total_Students)*100,3)
Overall_passing=pass_math=school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70),"Student ID"].count()
pass_O=round((Overall_passing/Total_Students)*100,3)

```

```python
#Creating a dataframe with the output 
District_Summary=pd.DataFrame({"Total Schools":[Total_School],"Total Students": Total_Students,
                               "Total Budget":Total_Budget,"Average Math Score":average_math,
                               "Average Reading Score":average_reading,"% Passing Math":pass_m,
                              "% Passing Reading":pass_r,"% Overall Passing":pass_O})
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Creating a Groupby relationship with school name
Grouped_School= school_data_complete.groupby(["school_name"])
# Storing key group by calculations into a variable
Total_Student1=Grouped_School["student_name"].count()
Total_Budget1=Grouped_School["budget"].mean()
A_Math=Grouped_School["math_score"].mean()
A_Reading=Grouped_School["reading_score"].mean()
Avg_S=Total_Budget1/Total_Student1
District=Grouped_School["type"].unique().str.get(0)

#Filtered further to get a list of students who met the passing criteria for Math, Reading and Overall
m_p=list((school_data_complete['math_score']>=70).groupby(school_data_complete['school_name']).sum())
r_p=list((school_data_complete['reading_score']>=70).groupby(school_data_complete['school_name']).sum())
o_p=list(((school_data_complete['reading_score']>=70)& 
         (school_data_complete['math_score']>=70)).groupby(school_data_complete['school_name']).sum())
#Got the total count of students per school in a list
total=list(Grouped_School["student_name"].count())

#Created Lists to include passing percentage by subject
math_p=[]
read_p=[]
Overall_p=[]

```

```python
#Used a for loop to calculating the passing percentage in each category and append the results into a loop

for x in range(len(o_p)):
    percent=(o_p[x]/total[x])*100
    Overall_p.append(percent)
for x in range(len(m_p)):
    percent=(m_p[x]/total[x])*100
    math_p.append(percent)
for x in range(len(r_p)):
    percent=(r_p[x]/total[x])*100
    read_p.append(percent)

```

```python
#Inputted results into a Data Frame using dictionaries

Sch_Summary_df=pd.DataFrame({"School Type":District,"Total Students":Total_Student1, 
                            "Total School Budget":Total_Budget1,"Per Student Budget":Avg_S,
                             "Average Math Score": A_Math,"Average Reading Score":A_Reading, 
                             "% Passing Math":math_p,"% Passing Reading":read_p,
                             "% Overall Passing":Overall_p})



Sch_Summary_df
```
